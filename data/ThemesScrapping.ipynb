```python
import requests
import time
# Constants
MAL_USERNAME = 'PlasmaFists'
JIKAN_API_URL = "https://api.jikan.moe/v4"
MAL_API_URL = "https://api.myanimelist.net/v2"
MAL_CLIENT_ID = '07583d4579b7f663d5589edf375b5a7d' 

# Function to get the user's anime list from MyAnimeList
def get_user_anime_list(username):
    headers = {
        "X-MAL-CLIENT-ID": MAL_CLIENT_ID
    }
    all_animes = []
    offset = 0
    limit = 100  # Define the number of items to retrieve per request

    while True:
        
        response = requests.get(f"{MAL_API_URL}/users/{username}/animelist", headers=headers, params={'limit': limit, 'offset': offset})
        
        if response.status_code == 200:
            data = response.json()
            anime_list = data.get('data', [])
            if not anime_list:
                break  # Exit if there are no more anime entries
            
            all_animes.extend(anime_list)  # Collect all anime entries
            offset += limit  # Increment the offset for the next request
            
            # Check for the presence of a next page
            if not data.get('paging', {}).get('next'):
                break  # Exit if there's no next page
        else:
            print(f"Failed to retrieve user anime list: {response.status_code} - {response.text}")
            break

    print(f"Retrieved {len(all_animes)} anime entries for {username}.")
    return all_animes

# Function to get anime themes from Jikan API
def get_anime_themes(mal_id):
    time.sleep(3)
    response = requests.get(f"{JIKAN_API_URL}/anime/{mal_id}/themes")
    if response.ok:
        data = response.json().get('data', {})
        openings = data.get('openings', [])
        endings = data.get('endings', [])
        
        return openings, endings
    else:
        print(f"Failed to retrieve themes for MAL ID {mal_id}: {response.status_code} - {response.text}")
    return [], []

# Main function to gather all theme songs
def get_all_songs(username):
    anime_list = get_user_anime_list(username)
    all_songs = []

    for anime in anime_list:
        mal_id = anime['node']['id']
        openings, endings = get_anime_themes(mal_id)
        
        # Adding openings to all_songs
        for opening in openings:
            all_songs.append({
                'anime': anime['node']['title'],
                'song': opening,
                'type': 'OP'  # Marking as opening
            })

        # Adding endings to all_songs
        for ending in endings:
            all_songs.append({
                'anime': anime['node']['title'],
                'song': ending,
                'type': 'ED'  # Marking as ending
            })

    return all_songs

# Execute the script
songs = get_all_songs(MAL_USERNAME)

# Print all songs
if songs:
    for song in songs:
        print(f"{song['anime']} - {song['song']} ({song['type']})")
else:
    print("No songs found.")
```

```python
import os

# Set of file extensions to include
valid_extensions = {".cs", ".js", ".jsx", ".ts", ".tsx"}

# Common folders to exclude (like libraries, build outputs)
exclude_dirs = {"node_modules", "bin", "obj", "dist", ".git", "__pycache__"}

def print_source_files(path, indent=""):
    try:
        entries = sorted(os.listdir(path))
        for entry in entries:
            full_path = os.path.join(path, entry)
            if os.path.isdir(full_path):
                if entry in exclude_dirs:
                    continue
                print(indent + "📁 " + entry)
                print_source_files(full_path, indent + "    ")
            elif os.path.splitext(entry)[1] in valid_extensions:
                print(indent + "📄 " + entry)
    except PermissionError:
        print(indent + "[Permission Denied]")

# Example usage:
root_path = r"C:\Users\vinee\Desktop\Work\Main Projects\Kiroku"  # Replace with your directory
print(f"Root: {root_path}")
print_source_files(root_path)

```
